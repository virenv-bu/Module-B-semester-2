# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Example monthly data (approximate, not exact from the book)
months = pd.date_range("2023-01-01", periods=12, freq="ME")
values = np.array([62, 61, 60, 59, 58, 57, 56, 55, 54, 58, 64, 71])

fig, ax = plt.subplots(figsize=(11, 5), dpi=130)

# 1) Base line in light gray
ax.plot(months, values, color="#B8B8B8", linewidth=2.5, zorder=1)

# 2) Highlight the last 3 months (preattentive focus)
ax.plot(months[-3:], values[-3:], color="#1F77B4", linewidth=4, zorder=3)
ax.scatter(months[-3:], values[-3:], color="#1F77B4", s=55, zorder=4)

# 3) Minimal styling (SWD look)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_color("#BBBBBB")
ax.tick_params(axis="y", left=False, labelleft=False)
ax.tick_params(axis="x", colors="#555555")
ax.grid(False)

# 4) Clear title + subtitle
ax.set_title(
    "Water Quality Trend Improves at Year End\nFocus: last three months show a clear increase",
    loc="left",
    fontsize=15,
    fontweight="bold",
    color="#222222",
    pad=16
)

# 5) Direct labels (avoid legend)
ax.text(
    months[-1], values[-1] + 1.2, "Dec peak",
    color="#1F77B4", fontsize=10, fontweight="bold", ha="right"
)
ax.text(
    months[1], values[1] - 2.0, "Earlier months",
    color="#8A8A8A", fontsize=9
)

# Optional source note
ax.text(
    0.0, -0.16, "Source: synthetic data for assignment recreation",
    transform=ax.transAxes, fontsize=8.5, color="#777777"
)

plt.tight_layout()
plt.show()
```
